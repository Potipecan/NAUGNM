```python
from common import *
%load_ext autoreload
%autoreload 2
sizes = [16, 4, 512]

```

```python
dim = 3000
x_train, y_train, x_test, y_test = load_dataset(dim)


models = []
for s in sizes:
    model = init_model(dim, s, 0.0, keras.regularizers.l2())
    model_name = f'model_{s}_l2_no-drop'
    train(model, x_train, y_train)
    models.append(model)
    model = init_model(dim, s, 0.5, None)
    model_name = f'model_{s}_no-reg_dropout'
    train(model, x_train, y_train)
    models.append(model)
```

```python
for model in models:
    print(model.name)
    test(model, x_test, y_test)
```

```python
for s in sizes:
    plot_results(f'model_3000_{s}', f'model_3000_{s}_l2', f'model_3000_{s}_do0.5')
```
