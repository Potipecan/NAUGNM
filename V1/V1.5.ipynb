```python
from common import *
%load_ext autoreload
%autoreload 2
sizes = [16, 4, 512]
import matplotlib
matplotlib.rcParams['figure.figsize'] = [10, 8]
```

```python
dim = 3000
x_train, y_train, x_test, y_test = load_dataset(dim)
models = []
for s in sizes:
    model = init_model(dim, s, 0.25, keras.regularizers.l2())
    train(model, x_train, y_train)
    models.append(model)
```

```python
for model in models:
    print(model.name)
    test(model, x_test, y_test)
```

```python
dim = 5000
x_train, y_train, x_test, y_test = load_dataset(dim)
models = []
for s in sizes:
    model = init_model(dim, s, 0.0)
    train(model, x_train, y_train)
    models.append(model)
    model = init_model(dim, s, 0.0, keras.regularizers.l2())
    train(model, x_train, y_train)
    models.append(model)
    model = init_model(dim, s, 0.25, None)
    train(model, x_train, y_train)
    models.append(model)
```

```python
for model in models:
    print(model.name)
    test(model, x_test, y_test)
```
